```python
def train_models(X_train, Y_train, X_val, Y_val, Kernel, Cs, gammas):
    if Kernel == "linear":
        file = open("validation_linear_3.txt", "w")
        for j in Cs:
            print("C = %f"%j);
            model_name = "model_linear_" + str(j);
            model = SVC(C=j, kernel=Kernel);
            model.fit(X_train,Y_train);
            score = (model.score(X_val, Y_val));
            file.write(model_name + " : Score: %f\n"%score);
        file.close();
    if Kernel == "rbf":
        file = open("validation_rbf_3.txt", "w");
        for i in Cs:
            print("C = %f"%i);
            for j in gammas:
                print("----> gamma = %f"%j);
                model_name = "model_rbf_"+str(i)+"_" + str(j);
                model = SVC(C=i, kernel = Kernel, gamma = j);
                model.fit(X_train, Y_train);
                score = (model.score(X_val, Y_val));
                file.write(model_name + " : Score: %f\n"%score);
        file.close();
    return;
        
```

```python
train_models(X_train, Y_train, X_val, Y_val, 'linear', Cs , gammas);
```
